```python
import pandas as pd
import matplotlib.pyplot as plt

# === Load data ===
df1 = pd.read_csv(r"C:\ReEDS\ReEDS-2.0_Input_Processing\state_policies\old and new data comparison\old ReEDS input\rps_fraction0.csv")
df2 = pd.read_csv(r"C:\ReEDS\ReEDS-2.0_Input_Processing\state_policies\outputs\rps_fraction.csv")
df3 = pd.read_csv(r"C:\ReEDS\ReEDS-2.0_Input_Processing\state_policies\outputs\intermediate outputs\rps_fraction_intermediate.csv")

# === Get list of states ===
states = df1['st'].unique()
states.sort()

# === Setup subplot grid ===
num_states = len(states)
cols = 3  # You can adjust this for layout
rows = -(-num_states // cols)  # Ceiling division

fig, axes = plt.subplots(rows, cols, figsize=(20, 3.5 * rows), sharex=True, sharey=True)
axes = axes.flatten()

# === Plot each state's data ===
for i, state in enumerate(states):
    ax = axes[i]
    state_df1 = df1[df1['st'] == state]
    state_df2 = df2[df2['st'] == state]
    state_df3 = df3[df3['st'] == state]
    ax.plot(state_df1['t'], state_df1['rps_all'], label='rps_fraction0.csv', marker='s', markersize=7, color='C0')
    ax.plot(state_df2['t'], state_df2['rps_all'], label='rps_fraction.csv', color='C3', linestyle='-', linewidth=3)
    ax.plot(state_df3['t'], state_df3['rps_all'], label='rps_fraction_intermediate.csv', marker='o', markersize=5, linestyle='None', color='C3')
    ax.set_title(state)
    ax.tick_params(labelbottom=True) 
    ax.grid(True)

# === Remove any unused subplots ===
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# === Shared labels and legend ===
fig.legend(['rps_fraction_old.csv', 'rps_fraction_new (piecewise interpolated).csv', 'rps_fraction_new (not interpolated).csv'],
           loc='upper center', ncol=3, bbox_to_anchor=(0.5, 0.99))
fig.suptitle("State RPS All Fraction Comparison", fontsize=18, y=1)

plt.tight_layout(rect=[0.03, 0.03, 1, 0.99])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CES data
df1 = pd.read_csv(r"C:\ReEDS\ReEDS-2.0_Input_Processing\state_policies\old and new data comparison\old ReEDS input\ces_fraction0.csv")
df2 = pd.read_csv(r"C:\ReEDS\ReEDS-2.0_Input_Processing\state_policies\outputs\ces_fraction.csv")
df3 = pd.read_csv(r"C:\ReEDS\ReEDS-2.0_Input_Processing\state_policies\outputs\intermediate outputs\ces_fraction_intermediate.csv")

# Rename column '*t' to 't' for clarity
df1 = df1.rename(columns={'*t': 't'})
df2 = df2.rename(columns={'*t': 't'})
df3 = df3.rename(columns={'*t': 't'})

# Get list of states
states = sorted(df1['st'].unique())

# Setup subplot grid
cols = 3
rows = -(-len(states) // cols)  # Ceiling division

fig, axes = plt.subplots(rows, cols, figsize=(20, 3.5 * rows), sharex=True, sharey=True)
axes = axes.flatten()

# Plot each state's CES data
for i, state in enumerate(states):
    ax = axes[i]
    df1_state = df1[df1['st'] == state]
    df2_state = df2[df2['st'] == state]
    df3_state = df3[df3['st'] == state]

    ax.plot(df1_state['t'], df1_state['Value'], label='ces_fraction0.csv', marker='s', markersize=7, color='C0')
    ax.plot(df2_state['t'], df2_state['Value'], label='ces_fraction.csv', color='C3', linestyle='-', linewidth=3)
    ax.plot(df3_state['t'], df3_state['Value'], label='ces_fraction_intermediate.csv', marker='o', markersize=5, linestyle='None', color='C3')
    ax.set_title(state)
    ax.tick_params(labelbottom=True)
    ax.grid(True)

# Remove unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Shared legend
fig.legend(['ces_fraction_old.csv', 'ces_fraction_new (piecewise interpolated).csv', 'ces_fraction_new (not interpolated).csv'], loc='upper center', ncol=2)

plt.tight_layout(rect=[0.03, 0.03, 1, 0.99])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# === Load the old and new hydrofrac_policy data ===
df_old = pd.read_csv(r"C:\ReEDS\ReEDS-2.0_Input_Processing\state_policies\old and new data comparison\old ReEDS input\hydrofrac_policy0.csv")
df_new = pd.read_csv(r"C:\ReEDS\ReEDS-2.0_Input_Processing\state_policies\outputs\hydrofrac_policy.csv")

# === Sort and merge by state for alignment ===
df_old = df_old.sort_values("st")
df_new = df_new.sort_values("st")
df_merged = pd.merge(df_old, df_new, on="st", how="inner", suffixes=("_old", "_new"))

# === Bar positions with added spacing ===
x = np.arange(len(df_merged['st'])) * 1.5
width = 0.6

# === RPS_All Plot with bar rim ===
fig_rps, ax_rps = plt.subplots(figsize=(18, 6))
ax_rps.bar(x - width/2, df_merged['RPS_All_old'], width=width, label='RPS_All (old in 2022)',
           color='C0', edgecolor='black', linewidth=0.6, zorder=3)
ax_rps.bar(x + width/2, df_merged['RPS_All_new'], width=width, label='RPS_All (new in 2023)',
           color='C3', edgecolor='black', linewidth=0.6, zorder=3)
ax_rps.set_title("Hydrofrac Policy - RPS_All Comparison")
ax_rps.set_xlabel("State")
ax_rps.set_ylabel("Fraction")
ax_rps.set_xticks(x)
ax_rps.set_xticklabels(df_merged['st'], rotation=90)
ax_rps.legend()
ax_rps.grid(True, axis='y', zorder=0)
plt.tight_layout()

# === CES Plot with bar rim ===
fig_ces, ax_ces = plt.subplots(figsize=(18, 6))
ax_ces.bar(x - width/2, df_merged['CES_old'], width=width, label='CES (old in 2022)',
           color='C0', edgecolor='black', linewidth=0.6, zorder=3)
ax_ces.bar(x + width/2, df_merged['CES_new'], width=width, label='CES (new in 2023)',
           color='C3', edgecolor='black', linewidth=0.6, zorder=3)
ax_ces.set_title("Hydrofrac Policy - CES Comparison")
ax_ces.set_xlabel("State")
ax_ces.set_ylabel("Fraction")
ax_ces.set_xticks(x)
ax_ces.set_xticklabels(df_merged['st'], rotation=90)
ax_ces.legend()
ax_ces.grid(True, axis='y', zorder=0)
plt.tight_layout()

plt.show()
```
